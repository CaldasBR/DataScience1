## Funções gerais para análise

```python
#Agrupa uma coluna do dataset e faz a CONTAGEM por alguma outra coluna
def conta_grp_generico(df,agrupar, metrica):
    grp = df.groupby([agrupar])[metrica].count()
    return grp

#Agrupa uma coluna do dataset e faz a MÉDIA por alguma outra coluna
def media_grp_generico(df,agrupar, metrica):
    grp = df.groupby([agrupar])[metrica].mean()
    return grp

#Agrupa uma coluna do dataset e faz a SOMA por alguma outra coluna
def soma_grp_generico(df,agrupar, metrica):
    grp = df.groupby(df,[agrupar])[metrica].sum()
    return grp

#Agrupa uma coluna do dataset e obtém o MÁXIMO de alguma outra coluna
def max_grp_generico(df,agrupar, metrica):
    grp = df.groupby([agrupar])[metrica].max()
    return grp

#Agrupa uma coluna do dataset e obtém o MÍNIMO de alguma outra coluna
def min_grp_generico(df,agrupar, metrica):
    grp = df.groupby([agrupar])[metrica].min()
    return grp

#Exibe os resultador de um dataset com 2 colunas em output de texto
def exibe_groupby(grp,texto):
    for x,i in zip(grp,grp.index):
        print('{} {} {}'.format(x,texto,i))
```

```python
#Cria função para exibição de gráficos de barras
def groupbyBarChat(df,agrupar, metrica, label_x, label_y, titulo):
    #Gera valores de y com base na coluna escolhida
    val_y = conta_grp_generico(df,agrupar, metrica)
    #Gera nome das colunas do eixo X em string
    val_x = val_y.index
    val_x_str = [str(y) for y in val_x]
    #Adiciona título do gráfico
    plt.title(titulo, fontsize=16)
    #Adiciona label nos eixos
    plt.xlabel(label_x,fontsize=12)
    plt.ylabel(label_y,fontsize=12)
    #Gera gráfico
    plt.bar(val_x_str, val_y)
    #Coloca labels nas colunas
    labels = [str(x) for x in val_y]
    for i in range(len(val_y)):
        plt.text(i,val_y[val_y.index[i]]+(val_y.max()*0.02), labels[i],ha='center')
    #Exibe o gráfico
    plt.show()
```

```python
#Cria função para exibição de gráficos de barras horizontais
def groupbyBarHChat(df,agrupar, metrica, label_x, label_y, titulo):
    #Gera valores de x com base na coluna escolhida
    val_x = media_grp_generico(df,agrupar, metrica)
    #Gera nome das colunas do eixo Y em string
    val_y = val_x.index
    val_y_str = [str(y) for y in val_y]
    #Adiciona título do gráfico
    plt.title(titulo, fontsize=16)
    #Adiciona label nos eixos
    plt.xlabel(label_x,fontsize=12)
    plt.ylabel(label_y,fontsize=12)
    #Gera gráfico
    y_pos = [i for i, _ in enumerate(val_x)]
    plt.barh(y_pos, val_x)
    plt.yticks(y_pos, val_y_str)
    #Coloca labels nas colunas
    labels = [str(x) for x in val_x]
    for i in range(len(val_y)):
        plt.text(val_x[val_x.index[i]]+(val_x.max()*0.02),i, labels[i],ha='center')
    #Exibe o gráfico
    plt.show()
```

```python
#Cria função para exibição de gráficos de pizza
def groupbyPieChat(df,agrupar, metrica, titulo):
    #Agrupa dados
    sizes = conta_grp_generico(df,agrupar, metrica)
    #Gera labels
    str_lst = sizes.index
    labels = [str(x) for x in str_lst]
    #Plota gráfico
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels, shadow=True, autopct='%1.1f%%') 
    ax1.axis('equal')
    plt.title(titulo, fontsize=16)
    plt.show()
```

```python
#Cria função para exibição de gráficos de barras empilhadas
def groupbyBarStkdChat(df,empilharpor,agrupar, metrica, label_x, label_y, titulo):
    #Gera lista de empilhamentos
    empilhar = eval('df.'+empilharpor+'.unique()')
    #Cria dataframes para cada empilhamento
    for i in range(len(empilhar)):
        if str(empilhar[i]) != 'nan':
            locals()['empilhar_%d' % i] = eval('df.query('+"'" + empilharpor + ' == "' + str(empilhar[i]) + '"'+"'"+')')
            #Gera valores de y com base na coluna escolhida dentro do empilhamento i
            locals()['val_y_%d' % i] = conta_grp_generico(locals()['empilhar_%d' % i],agrupar, metrica)
            #Gera nome das colunas do eixo X em string
            locals()['val_x_%d' % i]  = locals()['val_y_%d' % i].index
            locals()['val_x_%d_str' % i] = [str(y) for y in locals()['val_x_%d' % i]]
            #Plota empilhamento i
            if i == 0:
                plt.bar(locals()['val_x_%d_str' % i], locals()['val_y_%d' % i])
                bottom_append = locals()['val_y_%d' % i]
            else:
                plt.bar(locals()['val_x_%d_str' % i], locals()['val_y_%d' % i], bottom=bottom_append)
                bottom_append = bottom_append + locals()['val_y_%d' % i]
            #Coloca labels nas colunas
            labels = [str(x) for x in locals()['val_y_%d' % i]]
            for j in range(len(locals()['val_y_%d' % i])):
                plt.text(j,bottom_append[bottom_append.index[j]]-int(labels[j])/2, labels[j], ha='center')
            

    #Adiciona título do gráfico
    plt.title(titulo, fontsize=16)
    #Adiciona label nos eixos
    plt.xlabel(label_x,fontsize=12)
    plt.ylabel(label_y,fontsize=12)
    plt.legend(empilhar)
    
    #Exibe o gráfico
    plt.show()
```
